```python
# Importando bibliotecas
import pandas as pd  # processamento de dados, leitura e escrita de arquivos CSV (ex.: pd.read_csv)
import numpy as np  # álgebra linear e operações numéricas
import matplotlib.pyplot as plt  # criação de gráficos e visualizações
import seaborn as sns  # visualização de dados baseada no Matplotlib
sns.set(style="whitegrid") # Configuração do estilo dos gráficos
```

```python
df = pd.read_csv('dataset/McDonalds.csv')  # leitura de dados do arquivo CSV e armazenamento em um DataFrame
```

Estas visualizações fornecem informações sobre a tendência de várias métricas financeiras ao longo dos anos, correlações entre diferentes métricas, comparação de métricas de desempenho e análise de rácios financeiros.

```python
# Análise de Tendências
plt.figure(figsize=(12, 8))
plt.plot(df['Year'], df['Market cap ($B)'], marker='o', color='#1f77b4', label='Market Cap ($B)')
plt.plot(df['Year'], df['Revenue ($B)'], marker='o', color='#2ca02c', label='Revenue ($B)')
plt.plot(df['Year'], df['Earnings ($B)'], marker='o', color='#9467bd', label='Earnings ($B)')
plt.xlabel('Ano')
plt.ylabel('Valor (em $B)')
plt.title('Análise de Tendências')
plt.legend()
plt.grid(True)
plt.show()
```

A tendência da capitalização de mercado parece estar a aumentar ao longo dos anos, o que sugere um sentimento positivo dos investidores e perspectivas de crescimento. As receitas e os lucros revelaram uma tendência global crescente ao longo dos anos, com algumas flutuações. Isto indica que a empresa tem sido capaz de aumentar as suas receitas e resultados ao longo do tempo.

```python
# Análise de Correlação
plt.figure(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", center=0)
plt.title('Matriz de Correlação')
plt.show()
```

```python
# Comparação de Métricas de Desempenho
df[['Year', 'Revenue ($B)', 'Earnings ($B)', 'Market cap ($B)']].plot(x='Year', kind='bar', figsize=(12, 8))
plt.title('Comparação de métricas de desempenho')
plt.xlabel('Ano')
plt.ylabel('Valor (em $B)')
plt.grid(axis='y')
plt.show()
```

```python
# Análise de Ratios
df[['Year', 'P/E ratio', 'P/S ratio', 'P/B ratio']].plot(x='Year', marker='o', figsize=(12, 8))
plt.title('Análise de Ratios')
plt.xlabel('Ano')
plt.ylabel('Ratio')
plt.grid(True)
plt.show()
```

P/E ratio, P/S ratio, and P/B ratio: P/E ratio apresenta flutuações ao longo dos anos, mas geralmente permanece dentro de um intervalo razoável. P/S ratio Também apresenta flutuações, mas mostra uma tendência relativamente estável. P/B ratio Apresenta alguns valores negativos, indicando potenciais problemas com o cálculo do valor contabilístico.

```python
# Visualização das receitas, os ganhos e o dinheiro em caixa ao longo dos anos
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Revenue ($B)'], marker='o', color='green', label='Revenue')
plt.plot(df['Year'], df['Earnings ($B)'], marker='o', color='blue', label='Earnings')
plt.plot(df['Year'], df['Cash on Hand ($B)'], marker='o', color='orange', label='Cash on Hand')
plt.xlabel('Ano')
plt.ylabel('Montante ($B)')
plt.title('Receitas, lucros e dinheiro em caixa ao longo dos anos')
plt.grid(True)
plt.legend()
plt.show()
```

As disponibilidades registam algumas flutuações, mas, em geral, seguem uma tendência crescente ao longo dos anos. Este facto sugere que a empresa tem sido capaz de manter ou melhorar a sua posição de liquidez.

```python
# Gráfico de barras comparando o rendimento dos dividentos em diferentes anos
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Dividend Yield (%)'], color='purple')
plt.xlabel('Ano')
plt.ylabel('Rendimento de Dividendos (%)')
plt.title('Rendimento dos dividendos ao longo dos anos')
plt.grid(True)
plt.show()
```

```python
# Gráfico de dispersão que mostra a relação entre o P/E ratio e EPS
plt.figure(figsize=(10, 6))
plt.scatter(df['P/E ratio'], df['EPS ($)'], color='red')
plt.xlabel('P/E ratio') 
plt.ylabel('EPS ($)') 
plt.title('Relação entre o ratio P/E e o EPS')
plt.grid(True)
plt.show()
```

```python
# Calculando os níveis de despesas e as responsabilidades ao longo do ano
plt.figure(figsize=(10,6))
plt.plot(df['Year'],df['Total debt ($B)'], marker='o',color='green',label='debtlevels')
plt.plot(df['Year'],df['Total liabilities ($B)'],marker='o',color='orange',label='liabilities')
plt.xlabel('Ano')
plt.ylabel('Montante ($B)')
plt.title('Níveis de Dívidas e Passivos ao Longo do Ano')
plt.grid(True)
plt.legend()
plt.show()
```

A dívida total e o passivo total aumentaram ao longo dos anos, o que pode indicar um aumento da alavancagem e das obrigações financeiras.

Rendibilidade do capital próprio (ROE): O ROE mede a rentabilidade de uma empresa, revelando o lucro que a empresa gera com o dinheiro que os accionistas investiram.

Rácio de dívida para capital próprio: O rácio indica a proporção de capital próprio e de dívida utilizada para financiar as afirmações de uma empresa.

Fluxo de caixa livre (FCF): O FCF representa o dinheiro que uma empresa é capaz de gerar depois de ter desembolsado o dinheiro necessário para manter ou expandir a sua base de activos. É calculado como o fluxo de caixa operacional menos as despesas de capital.

```python
# Cálculo do ROE e do rácio de dívida sobre o capital próprio
df['ROE'] = df['Earnings ($B)']/df['Net assets ($B)']
df['Debt to Equity Ratio'] = df['Total debt ($B)']/df['Net assets ($B)']

df.head(5)
```

```python
# Projeção do ROE ao longo do ano
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['ROE'], marker='o', color='blue', label='ROE')
plt.xlabel('Ano')
plt.ylabel('ROE')
plt.title('Retorno sobre o Patrimônio Líquido (ROE) ao Longo dos Anos')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Traçar o rácio entre a dívida e o capital próprio ao longo dos anos
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Debt to Equity Ratio'], marker='o', color='green', label='Debt to Equity Ratio')
plt.xlabel('Ano')
plt.ylabel('Relação Dívida/Patrimônio Líquido')
plt.title('Índice de Endividamento sobre o Patrimônio Líquido ao Longo dos Anos')
plt.grid(True)
plt.legend()
plt.show()
```

Em conclusão, embora o conjunto de dados revele tendências positivas em termos de capitalização de mercado, receitas e lucros, existem flutuações e desafios em termos de rendibilidade, margem operacional e níveis de endividamento merecem uma análise mais aprofundada.

```python
# Tendência Geral dos Indicadores (Market Cap, Revenue, Earnings)
plt.figure(figsize=(12, 8))
plt.plot(df['Year'], df['Market cap ($B)'], marker='o', label='Market Cap ($B)')
plt.plot(df['Year'], df['Revenue ($B)'], marker='o', label='Revenue ($B)')
plt.plot(df['Year'], df['Earnings ($B)'], marker='o', label='Earnings ($B)')
plt.title('Tendência Geral dos Indicadores')
plt.xlabel('Ano')
plt.ylabel('Valores ($B)')
plt.legend()
plt.grid(True)
plt.show()
```

Ao longo dos anos analisados, o McDonald's demonstrou um crescimento consistente em sua capitalização de mercado e em seus lucros. Mesmo diante de flutuações econômicas, a empresa manteve uma trajetória ascendente, indicando resiliência e forte presença no mercado

```python
# Margem Operacional e EPS vs Lucro
plt.figure(figsize=(12, 8))
plt.plot(df['Year'], df['Operating Margin (%)'], marker='o', label='Operating Margin (%)')
plt.plot(df['Year'], df['EPS ($)'], marker='o', label='EPS ($)')
plt.plot(df['Year'], df['Earnings ($B)']*10, marker='o', label='Earnings ($B) * 10') # Multiplicado para escala
plt.title('Margem Operacional, EPS vs Lucro')
plt.xlabel('Ano')
plt.ylabel('Percentagem / EPS / Earnings ($B)')
plt.legend()
plt.grid(True)
plt.show()
```

A margem operacional e o lucro por ação (EPS) apresentaram uma correlação positiva com o lucro líquido. Isso sugere que o McDonald's conseguiu manter uma boa eficiência operacional, convertendo suas operações em lucros consistentes para os acionistas.

```python
# P/E Ratio
plt.figure(figsize=(12, 8))
plt.plot(df['Year'], df['P/E ratio'], marker='o', label='P/E Ratio')
plt.title('Relação P/E ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('P/E Ratio')
plt.grid(True)
plt.show()
```

O índice P/E variou ao longo dos anos, sugerindo diferentes expectativas do mercado quanto ao crescimento futuro do McDonald's. Picos no P/E podem indicar que o mercado esperava um crescimento significativo, mas também podem refletir uma possível sobrevalorização da empresa em determinados períodos.

```python
# Dívida em relação aos Ativos Totais
plt.figure(figsize=(12, 8))
plt.plot(df['Year'], df['Total debt ($B)'] / df['Total assets ($B)'], marker='o', label='Dívida/Ativos Totais')
plt.title('Dívida em relação aos Ativos Totais')
plt.xlabel('Ano')
plt.ylabel('Dívida/Ativos Totais')
plt.grid(True)
plt.show()
```

A relação entre dívida e ativos totais permaneceu relativamente estável ao longo dos anos, indicando uma alavancagem financeira controlada. Isso sugere que o McDonald's tem gerido sua estrutura de capital de forma prudente, embora o aumento gradual da dívida mereça atenção contínua.

```python
# Rendimento de Dividendos e Confiança do Investidor
plt.figure(figsize=(12, 8))
plt.plot(df['Year'], df['Dividend Yield (%)'], marker='o', label='Dividend Yield (%)')
plt.plot(df['Year'], df['Dividend (stock split adjusted) ($)'], marker='o', label='Dividend ($)')
plt.title('Rendimento de Dividendos e Dividendos Ajustados')
plt.xlabel('Ano')
plt.ylabel('Percentagem / Dividendos ($)')
plt.legend()
plt.grid(True)
plt.show()
```

Tanto o rendimento de dividendos quanto os dividendos ajustados por ação aumentaram consistentemente, refletindo a confiança dos investidores na sustentabilidade financeira do McDonald's e seu compromisso em retornar valor aos acionistas.

```python
# Capitalização de Mercado e Desempenho Financeiro
plt.figure(figsize=(12, 8))
plt.plot(df['Year'], df['Market cap ($B)'], marker='o', label='Market Cap ($B)')
plt.plot(df['Year'], df['Revenue ($B)'], marker='o', label='Revenue ($B)')
plt.plot(df['Year'], df['Earnings ($B)'], marker='o', label='Earnings ($B)')
plt.plot(df['Year'], df['Total assets ($B)'], marker='o', label='Total Assets ($B)')
plt.title('Capitalização de Mercado vs Desempenho Financeiro')
plt.xlabel('Ano')
plt.ylabel('Valores ($B)')
plt.legend()
plt.grid(True)
plt.show()
```

A capitalização de mercado do McDonald's acompanhou de perto as tendências de receita, lucro e ativos totais, sugerindo que o mercado avalia a empresa de forma consistente com seu desempenho financeiro.

```python
# Relação Valor de Mercado e Valor Contábil (P/B ratio)
plt.figure(figsize=(12, 8))
plt.plot(df['Year'], df['P/B ratio'], marker='o', label='P/B Ratio')
plt.title('Relação Valor de Mercado e Valor Contábil (P/B Ratio)')
plt.xlabel('Ano')
plt.ylabel('P/B Ratio')
plt.grid(True)
plt.show()
```

O P/B ratio do McDonald's apresentou variações, incluindo valores negativos em alguns anos, o que pode indicar desafios financeiros ou uma percepção de risco elevado. Isso sugere que, em determinados períodos, o mercado avaliou a empresa abaixo de seu valor contábil, possivelmente devido a incertezas ou dificuldades enfrentadas.
